## We load data

```python
filename = "kaggle-house-prices-train.csv"
json_filename = 'features_quali.json'
data = pd.read_csv(filename, index_col=0)
```

```python
def data_load():
    if path.exists(json_filename):
        with open(json_filename, 'r') as file:
            project_data = json.load(file)
        return project_data
    return {}
```

```python
json_feat = data_load()
pprint(json_feat)
```
